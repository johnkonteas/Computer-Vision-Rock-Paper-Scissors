```python
import cv2
import time
from keras.models import load_model
import numpy as np
from PIL import Image, ImageOps
from random import choice

print("\nYou are about to play rock, paper, scissors with the computer...." 
      "\n\nThe rules are simple and clear:"
        "\n\nRock smashes scissors."
        "\nScissors cut paper."  
        "\nPaper covers rock."
      "\n\nKeep playing until one of you has 3 wins !!!")

name = input("\nPlease give us your name  : ").capitalize()

while True:          
    cap = cv2.VideoCapture(0)
    data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
    c = 0  # counter for computer wins
    p = 0  # counter for player wins
    e = 0  # counter for empty inputs
    d = 0  # counter for draws
    
    while c != 3 and p != 3:
        
        p0 = choice(['rock', 'paper', 'scissors'])
        p1 = str(p0)
  
        # Load the model
        model = load_model('keras_model.h5', compile = 'False')

        while True:   
            # set the countdown timer
            timer = 3                                      
            ret, frame = cap.read()
            resized_frame = cv2.resize(frame, (224, 224), interpolation = cv2.INTER_AREA)
            image_np = np.array(resized_frame)
            normalized_image = (image_np.astype(np.float32) / 127.0) - 1 # Normalize the image
            data[0] = normalized_image
            prediction = model.predict(data)
            cv2.imshow('frame', frame)
            
            # PRESS q OR Q TO CAPTURE IMAGE !!
            
            if prediction[0][0] > 0.7:
                p2 = 'rock'
                print("rock")                            
            elif prediction[0][1] > 0.7:
                p2 = 'paper'
                print("paper")                                
            elif prediction[0][2] > 0.7:
                p2 = 'scissors'
                print("scissors")                
            else:
                p2 = ""
                print("nothing")
            
            
            # check for the key pressed
            k = cv2.waitKey(1)
                        
            # if key pressed is q
            if k == ord('q') or k == ord('Q'):
                                  
                prev = time.time()
 
                while timer >= 0:
                    ret, frame = cap.read()
 
                    # Display countdown on each frame
                    # specify the font and draw the
                    # countdown using puttext
                    font = cv2.FONT_HERSHEY_SIMPLEX
                    cv2.putText(frame, str(timer),
                        (200, 250), font,
                        7, (0, 255, 255),
                        4, cv2.LINE_AA)
                    cv2.imshow('frame', frame)
                    cv2.waitKey(1)
 
                    # current time
                    cur = time.time()
 
                    # if time elapsed is one second
                    # than decrease the counter
                    if cur-prev >= 1:
                        prev = cur
                        timer = timer-1
                           
                else:
                    ret, frame = cap.read()
 
                    # Display the clicked frame for 2 seconds                    
                    cv2.imshow('frame', frame)
 
                    cv2.waitKey(2000)                     
 
        #   Press Esc to exit
            elif  k == 27:
                break                         
                
        print('\nComputer : ', p1, f'\n{name} : ', p2)
        
        if p2 != "":
            p3 = str(p2)
            if p3 == 'rock' or p3 == 'paper' or p3 == 'scissors':
                if p1 == p3:
                    d += 1
                    print("\nIt's a draw!")
                elif p1 == "rock" and p3 == "scissors":
                    c += 1
                    print("\nComputer wins..")
                elif p1 == "paper" and p3 == "rock":
                    c += 1
                    print("\nComputer wins..")
                elif p1 == "scissors" and p3 == "paper":
                    c += 1
                    print("\nComputer wins..")
                else:
                    p += 1
                    print(f"\n{name} wins !!!")
        else:
            e += 1
            print("\nAh we probably got nothing as input, please retype !")


    print("\nTotal Computer wins   : ", c)
    print("Total Player wins     : ", p)
    print("Total Draws           : ", d)
    print("Empty Inputs          : ", e)


    if c > p:
        print("\nCOMPUTER IS THE FINAL WINNER :( ")
    else:    
        print(f"\n{name.upper()} IS THE FINAL WINNER !!! ")                                
    
    play_again = input("\nWanna play again? (y/n): ")
    if play_again.lower() != "y" and play_again.lower() != "yes":
        break
    else:
        # After the loop release the cap object
        cap.release()
        #Destroy all the windows
        cv2.destroyAllWindows()
```
